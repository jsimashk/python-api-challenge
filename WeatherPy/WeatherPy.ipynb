```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
print(api_key)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
'''
faanui: <Response [502]>
ostrovnoy: <Response [502]>
burkhala: <Response [502]>
khormuj: <Response [502]>
novyy urengoy: <Response [502]>
'''
url = "http://api.openweathermap.org/data/2.5/weather?"
city = "komsomolskiy" #'faanui'
url = url+"q="+city+"&APPID="+api_key+"&units=imperial"    
weather = requests.get(url)
print(weather)
if (weather.status_code == 200):
    print("success")
else:
    print("fail")
```

```python
# Build query URL and request your results in Celsius
url = "http://api.openweathermap.org/data/2.5/weather?"

weather_data = []
#time.sleep(60)
    
#API allows only 60 calls per minute so need to keep track of calls and add wait time
i = 0

for city in cities:
    
    if i >= 50:
        print("Waiting... Waiting... Waiting... ")
        time.sleep(100)
        i = 0

    url = url+"q="+city+"&APPID="+api_key+"&units=imperial"    
        
    weather = requests.get(url)
    i = i + 1
    status_code = weather.status_code
    
    
    if (status_code == 200):
        weather = weather.json()
        #i = i + 1
        
        weather_data.append({"city": city,
                         "clouds": weather["clouds"]["all"],
                         "country": weather["sys"]["country"],
                         "data": weather["dt"],
                         "humidity": weather["main"]["humidity"],
                         "latitude": weather["coord"]["lat"],
                         "longitude": weather["coord"]["lon"],
                         "max temperature": weather["main"]["temp_max"],
                         "wind speed": weather["wind"]["speed"]
                        })
        print(f"Status: {status_code} | Processing {i}...City {city} added to the list")
    else:
        print(f"Status: {status_code} | Skipping {i}.... City {city} not found")

        
cities_df = pd.DataFrame(weather_data)
cities_df.head()
```
